## Prepare the dataset

You will use the [Census Income dataset](https://archive.ics.uci.edu/ml/datasets/Adult) for Part 2 of the Homework 1 assignment. The task for this dataset is to predict whether an individual's income exceeds $50,000 per year or not based on a set of attributes/features (including age, occupation, education, and other factors). The dataset was constructed from the 1994 US Census database.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
seed = 0

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 
                'marital-status', 'occupation', 'relationship', 'race', 'sex', 
                'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
                'class']

data = data.drop(['fnlwgt', 'education-num'], axis=1)

data = data.replace(' ?', np.nan)

data.sample(10, random_state = seed)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data_dict = dict(data.isnull().sum())
null_cols = []
for column in data_dict:
  if data_dict[column]:
    null_cols.append(column)
null_cols    
```

```python
data.info()
```

```python
for column in null_cols:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
map = {
    ' <=50K' : 0,
    ' >50K' : 1,
}
data['class'] = data['class'].replace(map)
data
```

```python
categorical_columns = list(data.dtypes[data.dtypes == 'object'].index)
numeric_columns = list(data.dtypes[data.dtypes != 'object'].index)
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder() 

ordinal_encoded_data = oe.fit_transform(data[categorical_columns])
ordinal_encoded_data = pd.DataFrame(ordinal_encoded_data, columns = categorical_columns)

ordinal_encoded_data.head()
```

```python
encoded_data = pd.concat([data[numeric_columns], ordinal_encoded_data], axis = 'columns')
encoded_data.head()
```

```python
X = encoded_data.drop(['class'], axis=1)
y = encoded_data['class']
```

```python
X[X.columns] = (X[X.columns]-np.mean(X[X.columns], axis=0))/np.std(X[X.columns], axis=0)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .4, stratify = encoded_data['class'], random_state = 0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Model training

The details of training your model will vary depending on which model you choose to implement. For all models, you will find the optimal hyperparameters of your model using 5-fold cross validation and Grid Search.

```python
params = {'n_neighbors': list(range(1, 5)), 
              'weights': ['uniform', 'distance'],
              'metric': ['l1', 'l2', 'cosine']
             }
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
model = KNeighborsClassifier()

gridsearch = GridSearchCV(model, param_grid = params, cv = 5, verbose = 1, n_jobs = -1,return_train_score= True)
gridsearch.fit(X_train, y_train)
```

```python
gridsearch_results = pd.DataFrame.from_dict(gridsearch.cv_results_)
gridsearch_results.head()
```

```python
gridsearch.best_params_
```

```python
best_model = KNeighborsClassifier(**gridsearch.best_params_)
best_model.fit(X_train,y_train)
```

## Evaluation

Your final model evaluation should be performed on the test set. Report the following metrics: 
- Overall accuracy
- Precision
- Recall
- F1 score

```python
# YOUR CODE HERE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_best_model_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_best_model_pred)
precision = precision_score(y_test, y_best_model_pred)
recall = recall_score(y_test, y_best_model_pred)
f1 = f1_score(y_test, y_best_model_pred)
```

```python
print('The accuracy is  :{} %'.format(accuracy*100))
print('The precision is :{}'.format(precision))
print('The recall is    :{}'.format(recall))
print('The f1 score is  :{}'.format(f1))
```
